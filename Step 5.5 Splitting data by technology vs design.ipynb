```python
import pandas as pd
```

```python
final = pd.read_csv('data/final.csv')
final.head()
```

# mainclass

```python
#select the mainclass aggregation
final_main = final.drop(labels = ['cited_patent_number', 'cite_subclass','focal_subclass'], axis=1).drop_duplicates()
final_main.head()
```

```python
#design citing design
final_main_d2d = final_main.loc[final_main.cite_mainclass.str.match('D')]
final_main_d2d.head()
```

```python
final_main_d2u = final_main.loc[~final_main.cite_mainclass.str.match('D')]
final_main_d2u.head()
```

```python
#save
final_main_d2d.to_csv('data/final_main_d2d.csv', index_label=False)
final_main_d2u.to_csv('data/final_main_d2u.csv', index_label=False)
```

# Subclass

```python
final_sub = final.drop(labels = ['cited_patent_number', 'cite_mainclass','focal_mainclass'], axis=1).drop_duplicates()
final_sub.head()
```

```python
#design citing design
final_sub_d2d = final_sub.loc[final_sub.cite_subclass.str.startswith('D')]
final_sub_d2d.head()
```

```python
#design citing utility
final_sub_d2u = final_sub.loc[~final_sub.cite_subclass.str.startswith('D')]
final_sub_d2u.head()
```

```python
#save
final_sub_d2d.to_csv('data/final_sub_d2d.csv', index_label=False)
final_sub_d2u.to_csv('data/final_sub_d2u.csv', index_label=False)
```
